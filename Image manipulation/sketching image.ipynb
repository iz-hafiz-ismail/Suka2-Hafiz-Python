```python
import numpy as np
import imageio
import scipy.ndimage
```

```python
def dodge(front,back):
    result=front*255/(255-back) 
    result[result>255]=255
    result[back==255]=255
    return result.astype('uint8')
```

```python
def grayscale(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
```

```python
img ="https://avatars.githubusercontent.com/u/67820671?v=4"
```

```python
s = imageio.imread(img)
g = grayscale(s)
i = 255-g

b = scipy.ndimage.filters.gaussian_filter(i,sigma=6)
r= dodge(b,g)
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt
plt.imshow(r, cmap="gray")
```
