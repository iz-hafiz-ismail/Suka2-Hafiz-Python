```python
import qrcode

# to display images
from PIL import Image  
from IPython.display import display
```

```python
# basic
imgBasic = qrcode.make("link or text")
imgBasic.save('Directory\projectName.jpg', 'JPEG')

# Display Image
pil_im = Image.open('Directory\projectName.jpg')
display(pil_im)
```

```python
# advanced
# version parameter accept integer 1-40, use NONE to make system detect automatically (to use NONE, enable fit parameter)
qr = qrcode.QRCode(
    version=30,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4,
)
qr.add_data('text or link')
qr.make(fit=True)
imgAdvance = qr.make_image(fill_color="black", back_color="white")
imgAdvance.save('Directory\projectName2.jpg', 'JPEG')

# Display Image
pil_im = Image.open('Directory\projectName2.jpg')
display(pil_im)
```
