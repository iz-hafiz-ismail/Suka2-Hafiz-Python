```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.lyrics.com/lyric/37422084/Adam+Grahn/Running+With+The+Wolves')
print(page)
```

```python
html = BeautifulSoup(page.text, "html.parser")
print(html)
```

```python
lyrics = html.find("pre", class_="lyric-body").text
print(lyrics)
```
